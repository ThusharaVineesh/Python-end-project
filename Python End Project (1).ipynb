```python
salaryand_team=df.groupby('Team')['Salary'].sum()
highestsalary_team=salaryand_team.idxmax()
salaryand_position=df.groupby('Position')['Salary'].sum()
highestsalary_position=salaryand_position.idxmax()
print(f'Team which have highest salary expenditure: {highestsalary_team}')
print(f'Position which have highest salary expenditure: {highestsalary_position }')
```

```python
salaryand_team
```

```python
plt.figure(figsize=(60,8))
plt.bar(salaryand_team.index,salaryand_team)
plt.xlabel('Team')
plt.ylabel('Salary')
plt.title('Highest Salary Expenditure by team')
plt.show()
```

```python
#insight:Highest salary expentiure is in team Cleveland Cavaliers where as lowest is in Philadelphia 76ers 
```

```python
plt.bar(salaryand_position.index,salaryand_position)
plt.xlabel('Team')
plt.ylabel('Position')
plt.title('Highest Salary Expenditure by position')
plt.show()
```
